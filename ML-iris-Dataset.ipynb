```python
from sklearn.datasets import load_iris

iris=load_iris()
print(iris.keys())
print(iris["target_names"])

n_samples, n_features = iris.data.shape

print('Number of samples:' ,n_samples)
print('Number of features:' ,n_features)

print('Feature Name: ',iris.feature_names)
print('Target Name: ',iris.target_names)

print("Dimension of Input: ", iris.data.shape)
print("Dimension of Output: ", iris.target.shape)

print("First  rows of input: ",iris.data[1:5])
print("Target Data: ",iris.target)

```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()

x_index=3
colors=['blue','red','green']
for label, color in zip(range(len(iris.target_names)),colors):
  ax.hist(iris.data[iris.target==label,x_index],label=iris.target_names[label],color=color)
  ax.set_xlabel(iris.feature_names[x_index])
  ax.legend(loc='upper right')

  fig.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
fig,ax=plt.subplots()

x_index=3
y_index=0

colors=['blue','red','green']

for label,color in zip(range(len(iris.target_names)),colors):
  ax.scatter(iris.data[iris.target==label,x_index],iris.data[iris.target==label,y_index],
  label=iris.target_names[label],c=color)

  ax.set_xlabel(iris.feature_names[x_index])
  ax.set_ylabel(iris.feature_names[y_index])

  ax.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
n=len(iris.feature_names)

fig,ax=plt.subplots(n,n,figsize=(16,16))

colors=['blue','red','green']

for x in range(n):
  for y in range(n):
    xname=iris.feature_names[x]
    yname=iris.feature_names[y]
    for color_ind in range(len(iris.target_names)):
      ax[x,y].scatter(iris.data[iris.target==color_ind,x],iris.data[iris.target==color_ind,y],label=iris.target_names[color_ind],c=colors[color_ind])

      ax[x,y].set_xlabel(xname)
      ax[x,y].set_ylabel(yname)
      ax[x,y].legend(loc='upper left')
plt.show()
```

```python
import pandas as pd
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)

iris_df['Target']=iris.target
print(iris_df.head(5))
print(iris_df.info())
iris_df['Target']=iris_df['Target'].astype(object)
print(iris_df.info())
print(iris_df.groupby('Target').size())
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
wine=load_wine()

n=len(wine.feature_names)

fig,ax=plt.subplots(n,n,figsize=(30,30))

colors=['blue','red','green']

for x in range(n):
  for y in range(n):
    xname=wine.feature_names[x]
    yname=wine.feature_names[y]
    for color_ind in range(len(wine.target_names)):
      ax[x,y].scatter(wine.data[wine.target==color_ind,x],wine.data[wine.target==color_ind,y],label=wine.target_names[color_ind],c=colors[color_ind])

      ax[x,y].set_xlabel(xname)
      ax[x,y].set_ylabel(yname)
      ax[x,y].legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()

n = 3  # Select only 3 features for plotting
feature_indices = [0, 1, 2]  # Choosing first three features for visualization
fig, ax = plt.subplots(n, n, figsize=(15, 15))

colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'magenta']

for i in range(n):
    for j in range(n):
        x_index = feature_indices[i]
        y_index = feature_indices[j]

        for label in np.unique(digits.target):  # Loop through digits (0-9)
            mask = digits.target == label
            ax[i, j].scatter(digits.data[mask, x_index], digits.data[mask, y_index],
                             label=str(label), c=colors[label], alpha=0.5)

        ax[i, j].set_xlabel(f'Feature {x_index}')
        ax[i, j].set_ylabel(f'Feature {y_index}')
        ax[i, j].legend(loc='upper left', fontsize=8)

plt.show()
```
